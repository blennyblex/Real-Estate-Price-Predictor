```python
for i in data.drop("CHAS",1).columns:
    print("The skew for {} is {}".format(i, data[i].skew()))
```

```python
from scipy import stats

skew_feat = data.drop(["ZN","CHAS","MEDV"], 1)
unskew_feat = data[["ZN","CHAS","MEDV"]]

for i in skew_feat.columns:
    if skew_feat[i].skew() <-1 or skew_feat[i].skew() >1:
        print("The skew for {} is {}".format(i, skew_feat[i].skew()))
        
        skew_feat[i] = stats.boxcox(skew_feat[i])[0]
        
        print("The New skew for {} is {}\n".format(i, skew_feat[i].skew()))
```

```python
df = pd.concat([skew_feat, unskew_feat], 1)
df.head()
```

```python
X = df.drop(["MEDV", "CHAS"], 1)
y = df["MEDV"]
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler =StandardScaler()


x = pd.DataFrame(scaler.fit_transform(X), columns = X.columns )
X = pd.concat([x, df["CHAS"]], 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 4)  #or 4236574)
# 4236574
```

```python
def metrics(pred, model):
    mae = mean_absolute_error(y_test, pred)
    mse = mean_squared_error(y_test, pred)
    rmse = np.sqrt(mse)

    print("Accuracy of test:",r2_score(y_test,pred))
    print("\nThe mean_absolute_error is {} \nThe mean_squared_error is {} \nThe root_mean_squared is {}".format(mae, mse, rmse))
    

def modeling(regressor):
    model = regressor
    model.fit(X_train, y_train)
    model_pred = model.predict(X_test)

    metrics(model_pred, model)
```

```python
modeling(CatBoostRegressor())

# n_repeat of 10 and n_split 0f 2 to give rmse of 3.5
```

### BaggingRegressor & ExtraTreeRegressor

```python
extra_tree = ExtraTreeRegressor(random_state=0)

reg = BaggingRegressor(extra_tree, random_state=0).fit(X_train, y_train)
br_pred  = reg.predict(X_test)

metrics(br_pred, reg)
```

```python
from sklearn.svm import SVR
svr = SVR(kernel = "linear", degree = 1)
svr.fit(X_train, y_train)
svr_pred = svr.predict(X_test)

metrics(svr_pred, svr)
```

```python

reg = [LinearRegression(), RandomForestRegressor(random_state = 44), LGBMRegressor(), XGBRegressor()]

for i in reg:
    print("\n\nThe metrics for {} is \n".format(i))
    modeling(i)
    print("====================================================")
```

```python
rfc = RandomForestRegressor(random_state = 44)
rfc.fit(X_train, y_train)
rfc_pred = rfc.predict(X_test)

mae = mean_absolute_error(y_test, rfc_pred)
mse = mean_squared_error(y_test, rfc_pred)
rmse = np.sqrt(mse)
rmse
```
